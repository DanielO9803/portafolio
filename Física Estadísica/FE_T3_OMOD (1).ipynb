```python
# Importamos las librerías que vamos a utilizar.

import random as rd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Hacemos uso de las funciones en python para crear esta característica

def X(cam, paso): #La parte en que se crea la función con nombre X
    pasos = []  #Guardamos una lista vacía para usarla posteriormente
    
    for i in range(0, cam):# Creamos loop para iterar en los valors de entrada de la función
        x_ = rd.choices([-1,1], weights = (1/2, 1/2),k = paso)
        
        
        """Hacemos uso de la función dentro de los métodos de random
        donde el método rd.choice nos pedirá varios parámetros
        por lo cual le asignamos estos conforme a nuestro problema.
        Luego lo agregamos cada uno de estos a nuestra lista vacía y
        regresamos el valor para que tengamos un arreglo de este"""
        
        pasos.append(x_)
    return np.array(pasos) 
```

```python
# Hacemos el uso de este problema como ejemplo una de 20 pasos

X(1,20) # Llamamos la función

# Hace lo esperado la función la función
```

### Ejercicio 4

Creamos nuestras nuestras listas usando la función X creada.

```python
n_0, n_1, n_2, n_3 = X(10000,10), X(10000,100), X(10000,1000), X(10000,10000) 
```

```python
# Revisamos el tamaño de los arreglos.
print(len(n_0), len(n_1), len(n_2), len(n_3), sep=" , ")
```

Todas son de 10, 000 por lo que está correcto. 

```python
#Creamos una lista con cada uno de estos arreglos.

L = [n_0, n_1, n_2, n_3]

# Checamos el tamaño de la lista 

len(L)
```

```python
# Creamos nueva lista para agregar los resultados.

Prom_0 = []
for i in L:
    lis = []
    for j in i:
        promedio = j.mean()
        lis.append(promedio)
    Prom_0.append(lis)
Prom_0 = np.array(Prom_0)


# Hacemos las gráficas de lo anterior


l = ["n = 10", "n = 100", "n = 1000", "n = 1000"] # Creamos la lista con cada uno de estos pasos 

for i, j in enumerate(Prom_0): # El loop para ingresar las gráficas
    
    plt.figure(figsize = (25,15)) # Configuramos el tamaño de la gráfica
    plt.title("Promedio de la Lista") # Título de las gráficas
    plt.ylabel("Promedio Variable Estocástica") # Nombre del eje y
    plt.xlabel("Listas") # Nombre del eje x
    plt.plot(range(0,10000), j, color = "b", label = l[i]) #Configuramos las variables a aparecer.
    plt.ylim([-1,1]) # El límite de los ejes y
    plt.legend()
    plt.show()
```

### Ejercicio 6

```python
L_0 = X(10,10000)
L_1 = X(100,10000)
L_2 = X(1000,10000)
```

```python
L_3 = X(10000,10000)
```

```python
# Las juntamos
N = [L_0,L_1,L_2,L_3]

# Calculamos las varianzas

Var_N = []

for i in N:
    v = np.var(i, axis = 1)
    Var_N.append(v)
    
# Hacemos los datos para las gráficas.

l_L = ["L = 10", "L = 100", "L = 1000", "L = 10000", "L = 100000"]
y_r = [10,100,1000,10000,100000]
for i,j in enumerate(Var_N):
    plt.figure(figsize=(25,15))
    plt.ylim([0.998, 1])
    plt.xscale('log')
    plt.yscale('linear')
    plt.title("Grafícas por lista")
    plt.xlabel("Lista")
    plt.ylabel("Promedio Variable Estocastica")
    plt.plot(range(0,y_r[i]), j, c ="g",
             linewidth = 1.5, linestyle = "-", label=l_L[i])
    plt.legend(loc='center left')
    plt.show()

```

Como podemos ver y como vimos en clase, la varianza tiende a aumentar, aunque en valores pequeños pero lo hace. Lo cual es acorde a lo visto en clase, aumentar el número de listas nos hace que aumente la varianza. Al dejar fijo, lo que hacemos aumentando otros valores es que tendemos a 0 y aquí aumentando las listas aumenta la variación de este dato.
